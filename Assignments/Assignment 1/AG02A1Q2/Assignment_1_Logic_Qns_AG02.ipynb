### Q2b(i) Identify (with justification) two inference algorithms among tt_entails, resolution-refutation, forward chaining and backward chaining algorithms to solve this problem. [1 Marks]    



For this KB, we could only use tt_entails and resolution-refutation inference algorithms. Because there are clauses that cannot expressed as Horn Clauses, as there are more than one positive literal, that's why we cannot use forward and backward chaining.

### Q2b(ii) Select the best of these two algorithms.  Provide the rationale behind your selection. [1 Marks]

Hint: Try solving the problem using both algorithms; it should give you insight in terms of time complexity; Exact time complexity analysis is not needed as it is beyond the scope of this module. 



tt_entails algorithm would be the better choice. As we have 18 variables, for tt-entails, there are at most 2^18 models for every query,and since we have to check every variables, that means we have 18 queries, the total computational complexity will be O(18*2^18). But for resolution-refutation, the negation of the query will be add to the list of clauses, and then select two clauses to do the resolution and generate new clause. Since we have 18 variables, the maximum number of clauses it will generate is 3^18(3 situations for every variable in a clause: positive, negative, and non-exist), and for each query that is False, theoretically the number of generated clause will hit the maxium number, but in practice the function will hang, as the clauses generated are too much for the RAM. That's why we should choose tt_entails.
