# Create Markov chain based language model with corpus as input

- Investigate the impact of 'order' on the quality of the text generated, computational complexity and time complexity



```python
homc_ptb = MarkovChainLanguageModel(
    corpus=corpus,                           # corpus object
    order=1,                                # Markov order    
)

print(homc_ptb.generate_text(length=30, trigger_word=None))
```

```python
# Predict next word for a sequence
query_sentence = "the united states said that is not "
print("Predicted next word:", homc_ptb.predict_next_word(query_sentence))
```
